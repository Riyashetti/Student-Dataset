```python
import pandas as panda
df = panda.read_csv("student-por.csv")
```

```python
# finding the avg first

avg_score = df['G3'].mean()
avg_score

#finding the above avg
above_avg_students =df[df['G3'] > avg_score]
above_avg_students [['G3']]

#count the no of go avobe avg
above = (df['G3'] > avg_score).sum()
print("Above avg:",above)

```

```python
# find the top 5 students
df.sort_values(by='G1', ascending=False).head(5)[['G1','G2','G3']]
```

```python
#compare simple test mark with find mark
df[['G1','G2','G3']].mean()
```

```python
# find how many students passed the find eaxm
passed = (df['G3'] >=10).sum()
print("Passed:",passed)
```

```python
# find the students with 0 absences
df[df['absences'] == 0]
```

```python
#find the students with imprived from G1 to
df[df['G3']>df['G1']][['G1','G3']]
```

```python
# find the students with performed consistently amoudng g1 g2 g3
df[(df['G1']==df['G2'])&(df['G2']==df['G3'])][['G1','G2','G3']]
```

```python
# matplolib
#matplotlib tool is helpful for data visualization
#histogram

import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['G3'], bins = 10)
plt.xlabel("G3")
plt.ylabel("Frequency")
plt.title('Histogram of G3')
plt.show()
```

```python
df.columns
```

```python
# barchart
# avg find marks scored by boyes and  girls
avg_score_by_gender = df.groupby('sex') ['G3'].mean()
avg_score_by_gender
plt.figure()
avg_score_by_gender.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Final Mark')
plt.title('Average Final Mark by Gender')
plt.show()
```

```python
#scatter plt
plt.figure()
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("study time")
plt.ylabel("G3")
plt.title("Scatter Plot of G1 vs G3")
plt.show()
```

```python
# box plot
plt.figure()
plt.boxplot(df['G3'])
plt.ylabel("G3")
plt.title("spread of final scores")
plt.show()
```

```python
#line plots
avg_marks = df[['G1','G2','G3']].mean()
plt.figure()
plt.plot(avg_marks.index, avg_marks.values, marker = 'o')
plt.xlabel("Subject")
plt.ylabel("Average Mark")
plt.title("Average Marks by Subject")
plt.show()
```

```python
# with bar chart we will  find the number of students
# who failed and passed the final exam
pass_ = df['G3'] >= 10
plt.figure()
plt.bar(['pass','fail'],[pass_.sum(),(~pass_).sum()])
plt.ylabel("Number of students")
plt.title("Number of students with passed and failed")
plt.show()
```

```python
# histogram _ absence distribution
plt.figure()
plt.hist(df['absences'], bins = 10,color = 'skyblue',edgecolor = 'red')
plt.xlabel("absences")
plt.ylabel("Frequency")
plt.title('Histogram of absences')
plt.show()
# color option blue red green orange purple skyblue
```
